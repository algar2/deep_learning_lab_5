```python
import torch
import glob
from PIL import Image
from torch.utils.data import DataLoader
from torchvision.utils import draw_bounding_boxes
from torchvision.models.detection import fasterrcnn_resnet50_fpn, FasterRCNN_ResNet50_FPN_Weights

weights = FasterRCNN_ResNet50_FPN_Weights.DEFAULT

# Getting classes of FasterRCNN
classes = weights.DEFAULT.meta["categories"]


transformations  = weights.transforms()



# Transforming images
image_dir = "sample_images/*.jpg"
image_files = glob.glob(image_dir)

images = []
for file_path in image_files:
    img = Image.open(file_path)
    images.append(img)

test_images = [transformations(img) for img in images]


model_rcnn = fasterrcnn_resnet50_fpn(weights)
model_rcnn.eval()

outputs = model_rcnn(test_images)

```

```python
import numpy as np 
import matplotlib.pyplot as plt
import torchvision.transforms.functional as F
from torchvision.io import decode_image

def get_class_name(class_id):
    return classes[class_id]

def draw_boxes(image, outputs, idx, fig_size=(10,10)):

    boxes = outputs[idx]['boxes'].detach().cpu().numpy()
    labels = outputs[idx]['labels'].detach().cpu().numpy()
    scores = outputs[idx]['scores'].detach().cpu().numpy()

    plt.figure(figsize=fig_size)

    image = np.array(image)
    for box, label, score in zip(boxes,labels,scores):
        x_min, y_min, x_max, y_max = box
        class_name = get_class_name(label)
        plt.imshow(image)
        plt.gca().add_patch(plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                            linewidth=2, edgecolor='r', facecolor='none'))
        plt.text(x_min, y_min, f"{class_name} ({score:.2f})", color='y')
    
    plt.axis('off')
    plt.show()

draw_boxes(Image.open("sample_images/toothbrushes_1.jpg"), outputs, 0)

# img.show("sample_images/toothbrushes_1.jpg")
# image_ints = [
# for image in test_images:
#     decoded = decode_image(image)
#     image_ints.append(decoded)

```
